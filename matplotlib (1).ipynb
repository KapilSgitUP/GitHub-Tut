```python
import numpy as np
```

```python
x=np.arange(0,5,0.1)
y=x**3
```

```python
x2=np.arange(0,5,0.1)
y2=x2**2
```

```python
print(x)
print(y)
```

```python
plt.plot(x,y,color="red",marker="o")
plt.show()
```

```python
plt.plot(x,y,linewidth=3,label="x**3")
plt.plot(x2,y2,label="x**2")
plt.axis([0,10,0,200])
plt.title("Matplotlib basics")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid()
plt.show()
```

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,420,180,250,300,370]
c_arr=np.arange(len(year))
# c_label=np.arange(len(year))
```

```python
c_arr
```

```python
c_arr=np.array(["black","orange","blue","red","grey","green"])
```

```python
plt.scatter(year ,salary,s=population,c=c_arr,alpha=1,edgecolor="black",marker="*")
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year ,salary,s=population, c=c_arr,alpha=1,edgecolor="black",marker="*")
plt.title("Bubble Chart")
plt.show()
```

```python
# pie chart
labels=["A","B","C","D"]
sizes=[3,4,6,2]
explode=[0.02,0.02,0.02,0.02]
colors=["blue","pink","orange","green"]
plt.pie(sizes,colors=colors,labels=labels,autopct="%.2f%%",explode=explode)
plt.show()
```
